# CORRELATION

Correlation is a statistical measure that indicates the degree to which two or more variables are related or associated. It quantifies the strength and direction of the linear relationship between two variables. In other words, correlation measures how changes in one variable are related to changes in another variable.

The correlation coefficient is the common metric used to express the correlation between two variables. It ranges between -1 and +1:

A correlation coefficient of +1 indicates a perfect positive correlation, meaning that when one variable increases, the other increases proportionally.
A correlation coefficient of -1 indicates a perfect negative correlation, meaning that when one variable increases, the other decreases proportionally.
A correlation coefficient close to 0 suggests a weak or no linear correlation between the variables.

# calculating correlation matrix using kendall


```python
# calculate correlation matrix
## using non-parametric test of correlation (kendall), since some features are binary
corrs = df.select_dtypes(include=np.number).corr(method = 'kendall')

# plot clustered heatmap of correlations
sns.clustermap(corrs, cmap='coolwarm', center=0);
```

# calculating correlation matrix using SPEARMAN

```python
# calculate correlation matrix
## using non-parametric test of correlation (kendall), since some features are binary
corrs = df.select_dtypes(include=np.number).corr(method = 'spearman')

# plot clustered heatmap of correlations
sns.clustermap(corrs, cmap='coolwarm', center=0);
```

### * From the Correlation matrix , it is found that the DIFFERENT AGE GROUPS has the highest positive correlation followed by TOTAL BOOSTERS AND VACCINATIONS.

# PLOTTING USING PIE CHART

```python
country_cases = df.groupby('location')['total_cases'].max().reset_index()
country_cases = country_cases.sort_values(by='total_cases', ascending=False)
top_countries = country_cases.head(10)

plt.figure(figsize=(10, 6))
plt.pie(top_countries['total_cases'], labels=top_countries['location'],
        autopct='%1.1f%%', startangle=140, colors=plt.cm.Set3.colors)
plt.title('Top 10 Countries with the Most COVID-19 Cases')
plt.axis('equal')
plt.show()
```
